```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_trace = list()
y_trace = list()
y_GT = list()

with open("out.txt", 'r') as file:
    for idx, line in enumerate(file):
        x, y = list(map(float, line.strip().split(',')))
        x_trace.append(x)
        y_GT.append(np.arctan(x*0.001)*180/np.pi)
        y_trace.append(y)
```

```python
plt.scatter(x_trace, y_trace, s= 3)
plt.plot(x_trace, y_GT, 'b', linewidth=1)

plt.grid(True)
plt.show()

error = 0
for y, y_gt in zip(y_trace, y_GT):
    error += np.abs(y-y_gt)

print(error)
```
