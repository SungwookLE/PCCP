```python
import matplotlib.pyplot as plt
```

```python
x_trace = list()
y_trace = list()

with open("tree.txt", 'r') as file:
    for line in file:
        x, y = map(float, line.strip().split(','))
        x_trace.append(x)
        y_trace.append(y)

x_sampling = list()
y_sampling = list()

with open("sampling.txt", 'r') as file:
    for line in file:
        x, y = map(float, line.strip().split(','))
        x_sampling.append(x)
        y_sampling.append(y)
```

```python
obj_x = list()
obj_y = list()

with open("object.txt", 'r') as file:
    for line in file:
        x, y = map(float, line.strip().split())
        obj_x.append(x)
        obj_y.append(y)
```

```python
plt.plot(x_trace, y_trace, color='r')
plt.plot(x_sampling, y_sampling, color='k', linewidth=1, alpha=0.3, marker='.')
plt.scatter(obj_x, obj_y, color='yellowgreen', label='Obstacle', s=100)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```
